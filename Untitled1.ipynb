```python




num_pairs = len(ustun)
cols = 3  # Number of columns in the subplot grid
rows = -(-num_pairs // cols)  # Ceiling division to determine number of rows

fig, axes = plt.subplots(rows, cols, figsize=(15, 30))  # Adjust the figure size as needed

for i, col in enumerate(ustun):
    ax = axes[i // cols, i % cols] if num_pairs > 1 else axes  # Select the subplot
    ax.plot(df['Time Serie'], df[col])
    ax.set_title(col)
    ax.set_xlabel('Time Serie')
    ax.set_ylabel('Value')

# Hide empty subplots
for i in range(num_pairs, rows * cols):
    ax = axes[i // cols, i % cols] if num_pairs > 1 else axes
    ax.axis('off')

plt.tight_layout()
plt.show()
```
